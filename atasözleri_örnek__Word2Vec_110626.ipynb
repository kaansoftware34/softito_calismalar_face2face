```python
d0 = "Damlaya damlaya göl olur."
d1 = "Ayağını yorganına göre uzat."
d2 = "Sakla samanı gelir zamanı."
d3 = "Bir elin nesi var, iki elin sesi var."
d4 = "Akıl akıldan üstündür."
d5 = "Gülü seven dikenine katlanır."
d6 = "Ağaç yaşken eğilir."
d7 = "Üzüm üzüme baka baka kararır."
d8 = "Acele işe şeytan karışır."
d9 = "Komşu komşunun külüne muhtaçtır."
korpus = [d0, d1, d2, d3, d4, d5, d6, d7, d8, d9]
```

```python
try:
    from gensim.models import Word2Vec
    from gensim.utils import simple_preprocess
    print("Gensim yüklü. Devam ediliyor...")
except ImportError:
    print("Gensim yüklü değil. Yükleniyor...")
    import sys
    !{sys.executable} -m pip install gensim -q
    from gensim.models import Word2Vec
    from gensim.utils import simple_preprocess
    print("Gensim yüklendi!")
```

```python
# Korpusu tokenize et
tokenize_doc = [simple_preprocess(doc) for doc in korpus]

# Word2Vec modelini eğit
w2v_model = Word2Vec(
    sentences=tokenize_doc,
    vector_size=30,          # vektör boyutu
    window=3,                # komşu penceresi
    min_count=2,             # minimum kelime frekansı
    sg=0,                    # 0=CBOW, 1=Skip-gram
    epochs=20,
    seed=42
)

import numpy as np # Import numpy

print(f"Kelime dağarcığı: {len(w2v_model.wv)}")
print(f"Vektör boyutu: {w2v_model.wv.vector_size}")
print(f"\n'damlaya' vektörü (ilk 10 değer): {w2v_model.wv['damlaya'][:10]}")
print(f"Vektör normu: {np.linalg.norm(w2v_model.wv['damlaya']):.4f}")
```

```python
# Benzer kelimeleri bulma
print("EN YAKIN KELİMELER\n")

for kelime in ["türk", "atatürk", "ey", "bir"]:
    print(f"'{kelime}' -> en benzer 5 kelime:")
    try:
        benzer = w2v_model.wv.most_similar(kelime, topn=5)
        for k, skor in benzer:
            print(f"    {k:12s} (skor: {skor:.4f})")
    except:
        print(f"    (kelime bulunamadı)")
    print()
```

```python
# Korpusu tokenize et
tokenize_doc = [simple_preprocess(doc) for doc in korpus]

# Word2Vec modelini eğit
w2v_model = Word2Vec(
    sentences=tokenize_doc,
    vector_size=30,          # vektör boyutu
    window=3,                # komşu penceresi
    min_count=2,             # minimum kelime frekansı
    sg=1,                    # 0=CBOW, 1=Skip-gram
    epochs=20,
    seed=42
)

import numpy as np # Import numpy

print(f"Kelime dağarcığı: {len(w2v_model.wv)}")
print(f"Vektör boyutu: {w2v_model.wv.vector_size}")
print(f"\n'damlaya' vektörü (ilk 10 değer): {w2v_model.wv['damlaya'][:10]}")
print(f"Vektör normu: {np.linalg.norm(w2v_model.wv['damlaya']):.4f}")
```

```python
# Benzer kelimeleri bulma
print("EN YAKIN KELİMELER\n")

# Korpus değiştiği için yeni kelime dağarcığındaki kelimeleri arıyoruz
aranacak_kelimeler = ["damlaya", "elin", "var", "baka"]

for kelime in aranacak_kelimeler:
    print(f"'{kelime}' -> en benzer kelimeler:")
    try:
        # Dağarcık küçük olduğu için topn=3 yapıyoruz
        benzer = w2v_model.wv.most_similar(kelime, topn=3)
        for k, skor in benzer:
            print(f"    {k:12s} (skor: {skor:.4f})")
    except:
        print(f"    (kelime bulunamadı)")
    print()
```
